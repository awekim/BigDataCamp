```python
from langchain.embeddings import HuggingFaceBgeEmbeddings
```

```python
model_name = "BAAI/bge-small-en-v1.5"
model_kwargs = {'device': 'cpu'}
encode_kwargs = {'normalize_embeddings': True}
hf = HuggingFaceBgeEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

```python
embedding_en_examples = {
    "todays is monday",
    "weather is nice today",
    "what's the problem",
    "langchain in useful",
    "Hello World",
    "my name is morris"
}

embedding_en = hf.embed_documents(embedding_en_examples)
```

```python
len(embedding_en)
```

```python
len(embedding_en[0])
```

```python
embedding_q = hf.embed_query("Hello? who is this?")
embedding_a = hf.embed_query("Hi this is harrison")
```

```python
print(cos_sim(embedding_q, embedding_a))
print(cos_sim(embedding_q, embedding_en[1]))
print(cos_sim(embedding_q, embedding_en[5]))
```

```python
model_name = "jhgan/ko-sbert-nli"
model_kwargs = {'device': 'cpu'}
encode_kwargs = {'normalize_embeddings': True}

ko = HuggingFaceBgeEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

```python
embedding_ko_examples = {
    "안녕하세요",
    "제 이름은 홍길동 입니다.",
    "이름이 무엇인가요",
    "랭체인은 유용합니다.",
    "홍길동 아버지의 이름은 홍상직 입니다."
}

embedding_ko = ko.embed_documents(embedding_ko_examples)
```

```python
embedding_q = ko.embed_query("홍길동은 아버지를 아버지라 부르지 못하였습니다. 홍길동 아버지의 이름은 무엇입니까?")
embedding_a = ko.embed_query("홍길동 아버지는 엄했습니다.")
```

```python
print(cos_sim(embedding_q, embedding_a))
print(cos_sim(embedding_q, embedding_ko[1]))
print(cos_sim(embedding_q, embedding_ko[3]))
print(cos_sim(embedding_q, embedding_ko[4]))
```
